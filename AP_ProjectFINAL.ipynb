```python
import numpy as np
import pandas as pd
```

```python
data=pd.read_csv("/content/CSVdata_final.csv")
data.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
data.shape
```

```python
colName=['Slot_L_move','Slot_r_move','W_slot_L','L_slot_L','L_slot_R']
x = data[colName]
x.head()
```

```python
y=data.drop(columns=colName,axis=1)
y.head()
```

```python
from keras.models import Sequential
from keras.layers import Dense, BatchNormalization
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(x, y, test_size=0.2, random_state=42)
model = Sequential()
model.add(Dense(32, input_dim=5, activation='relu'))
model.add(BatchNormalization())
model.add(Dense(64, activation='relu'))
model.add(BatchNormalization())
model.add(Dense(128, activation='relu'))
model.add(BatchNormalization())
model.add(Dense(256, activation='relu'))
model.add(BatchNormalization())
model.add(Dense(512, activation='relu'))
model.add(BatchNormalization())
model.add(Dense(1001, activation='linear'))

# Compile the model
model.compile(loss='mean_squared_error', optimizer='adam')

# Train the model
model.fit(X_train, Y_train, epochs=50, batch_size=10)

# Evaluate the model
loss = model.evaluate(X_test, Y_test)
print('Model Loss:', loss)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from math import sqrt

# Make predictions on the test set
predictions = model.predict(X_test)

# Calculate Mean Absolute Error (MAE)
mae = mean_absolute_error(Y_test, predictions)
print('Mean Absolute Error:', mae)

# Calculate Mean Squared Error (MSE)
mse = mean_squared_error(Y_test, predictions)
print('Mean Squared Error:', mse)

# Calculate Root Mean Squared Error (RMSE)
rmse = sqrt(mse)
print('Root Mean Squared Error:', rmse)

# Calculate R-squared (R²)
r2 = r2_score(Y_test, predictions)
print('R-squared:', r2)
```

```python
import matplotlib.pyplot as plt

# Generate equally spaced frequency values between 1 and 5 GHz
frequencies = np.linspace(1, 5, 1001)  # 1001 points between 1 and 5

# Make predictions
new_data = pd.DataFrame([[0.2,10.5,1.55,8,13]])  # insert new data here
predictions = model.predict(new_data)

# Plot the predictions
plt.figure(figsize=(10, 6))
plt.plot(frequencies, predictions[0], label='Predicted S11')
plt.xlabel('Frequency (GHz)')
plt.ylabel('S11 Parameter')
plt.title('Predicted S11 Parameter vs Frequency')
plt.legend()
plt.grid(True)
plt.show()
```
